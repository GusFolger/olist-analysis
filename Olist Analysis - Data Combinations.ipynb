# Combinations
# olist_order_customer_dataset
Combined from olist_customers_dataset and olist_orders_dataset

```python
# Merging df_customers on df_orders
df_ord_cust = df_customers.merge(df_orders, on = 'customer_id', indicator = True)
```

```python
# Checking dataframe head
df_ord_cust.head()
```

```python
# Removing unncessary columns
df_ord_cust = df_ord_cust.drop('Unnamed: 0_x', axis=1)
df_ord_cust = df_ord_cust.drop('Unnamed: 0_y', axis=1)
df_ord_cust = df_ord_cust.drop('_merge', axis=1)
```
